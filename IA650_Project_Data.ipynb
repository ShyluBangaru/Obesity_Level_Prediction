Chi-Squared Test

```python
import pandas as pd
from scipy.stats import chi2_contingency

df= pd.DataFrame(obesity_data)
categorical_vars = ['Gender', 'CALC']
results = {}

for var in categorical_vars:
    contingency_table = pd.crosstab(df[var], df['NObeyesdad'])
    chi2, p, dof, ex = chi2_contingency(contingency_table)
    results[var] = {'chi2': chi2, 'p-value': p, 'dof': dof, 'expected': ex}

for var, result in results.items():
    print(f"Variable: {var}")
    print(f"Chi-squared: {result['chi2']}")
    print(f"P-value: {result['p-value']}")
    print(f"Degrees of freedom: {result['dof']}")
    print(f"Expected frequencies: \n{result['expected']}\n")
```

F-Statistics

```python
from scipy.stats import f_oneway

continuous_cols = ['Age', 'Height', 'Weight', 'CH2O', 'FAF', 'TUE', 'NCP']
anova_results = {}

for col in continuous_cols:
    groups = [df[df['NObeyesdad'] == level][col] for level in df['NObeyesdad'].unique()]
    f_statistic, p_value = f_oneway(*groups)
    anova_results[col] = {'F-statistic': f_statistic, 'p-value': p_value}

for col, result in anova_results.items():
    print(f"Variable: {col}")
    print(f"F-statistic: {result['F-statistic']}")
    print(f"P-value: {result['p-value']}")
```

ANOVA

```python

import pandas as pd
import numpy as np
from scipy.stats import f_oneway
from scipy.stats import f

continuous_cols = ['Age', 'Height', 'Weight', 'CH2O', 'FAF', 'TUE', 'NCP']
categorical_col = 'NObeyesdad'

overall_mean = df[continuous_cols].values.mean()

SSTr = 0
SSE = 0
SST = 0

for level in df[categorical_col].unique():
    group = df[df[categorical_col] == level][continuous_cols].values
    group_mean = group.mean()
    group_size = len(group)
    
    SSTr += group_size * ((group_mean - overall_mean) ** 2)
    SSE += np.sum((group - group_mean) ** 2)
    SST += np.sum((group - overall_mean) ** 2)

df_between = len(df[categorical_col].unique()) - 1
df_within = len(df) - len(df[categorical_col].unique())
df_total = len(df) - 1

MSTr = SSTr / df_between
MSE = SSE / df_within

F = MSTr / MSE
p_value = 1 - f.cdf(F, df_between, df_within)

# Creating ANOVA table
anova_table = pd.DataFrame({
    'Source of Variation': ['Treatments', 'Errors', 'Total'],
    'SS': [SSTr, SSE, SST],
    'df': [df_between, df_within, df_total],
    'MS': [MSTr, MSE, ''],
    'F': [F, '', ''],
    'p-value': [p_value, '', '']
})

print(anova_table)
```

Scree Plot

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

numerical_features = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']

x = obesity_data[numerical_features]
x = StandardScaler().fit_transform(x)

pca = PCA()
pca.fit(x)

explained_variance_ratio = pca.explained_variance_ratio_

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.xticks(range(1, len(explained_variance_ratio) + 1))
plt.grid(True)
plt.show()
```

Interaction Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

seed = 42

sampled_data = obesity_data.sample(frac=0.1, random_state=seed)

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.pointplot(x='Age', y='Weight', hue='Gender', data=sampled_data, dodge=True, markers=['o', 's'], linestyles=['-', '--'])
plt.title('Interaction Plot: Age and Gender on Weight')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.legend(title='Gender')
plt.show()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np

formula = 'Weight ~ Age + Height + FCVC + NCP + CH2O + FAF + TUE + Gender'

model = smf.ols(formula=formula, data=obesity_data).fit()

main_effects = model.params[1:] 

plt.figure(figsize=(10, 6))
plt.bar(main_effects.index, main_effects.values)
plt.title('Main Effect Plot')
plt.xlabel('Variables')
plt.ylabel('Effect on Weight')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
